# Question 1

Plot using seaborn.relplot, line plot, box plot on fmri data set

```python
#Importing libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib as mpl 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#Available data set
sb.get_dataset_names()
```

```python
# Importing the Data set (fmri)
df=sb.load_dataset('fmri')
df.head(10)
```

```python
#Describing the type of data in columns
df.dtypes
```

**Line plot**

```python
# Ploting line plot
sb.set(style='darkgrid')
sb.relplot(x='timepoint',y='signal', hue='subject', size='region', style='event', data=df, legend='auto', kind='line', height=7, aspect=1,sort=True)
```

**Box plot**

```python
sb.set(style='darkgrid')
sb.boxplot(x='timepoint', y='signal', hue='event', data= df)
```

**Relational plot**

```python
sb.set(style='darkgrid')
sb.relplot(x='timepoint', y='signal', hue='subject', size='region', style='event', data=df, legend='auto', kind='scatter', height= 8, aspect=1)
```
